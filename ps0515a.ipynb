```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def numerical_derivative(fx, a, b, n):
    dfx = np.zeros(n+1)
    dx = (b-a)/n
    for i in range(1,n):
        dfx[i] = (fx[i+1] - fx[i-1])/(2*dx)
    
    dfx[0] = (-3*fx[0] + 4*fx[1] -fx[2])/(2*dx)
    dfx[n] = ( 3*fx[n] - 4*fx[n-1] +fx[n-2])/(2*dx)
    return dfx
```

```python
a = -1.0
b = 1.0
n = 2**18
x = np.arange(a,b+(b-a)/n,(b-a)/n)
fx = 1+np.sin(np.pi*(x+0.25))
x2 = np.arange(a,b+(b-a)/(2*n),(b-a)/(2*n))
fx2 = 1+np.sin(np.pi*(x2+0.25))
x4 = np.arange(a,b+(b-a)/(4*n),(b-a)/(4*n))
fx4 = 1+np.sin(np.pi*(x4+0.25))
plt.figure()
plt.plot(x, fx, x, numerical_derivative(fx, a, b, n), 'g', x, np.pi*np.cos(np.pi*(x+0.25)), 'r--' )
plt.grid()
plt.figure()
dfx2 = numerical_derivative(fx2, a, b, 2*n)
dfx4 = numerical_derivative(fx4, a, b, 4*n)
plt.plot(x, numerical_derivative(fx, a, b, n)- dfx2[::2],'r-.', x, dfx2[::2]-dfx4[::4],'c-.')
plt.grid()
```
